```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a, b = 998, 1998
c, d= -999, -1999
h = 1 / 1000
X = 1
x = np.arange(0, X, h)
u0, V0 = -3, 5
solution = np.zeros((x.size, 2))
solution[0] = np.array([u0, v0])
f = lambda _u, _v: np.array([_u * a + b * _v, _u * c + d * _v])
```

```python
for i in range(x.size - 1):
    solution[i + 1] = solution[i] + h * f(*solution[i])
u = solution[:, 0]
v = solution[:, 1]
plt.scatter(x, u, c='b', label='u')
plt.scatter(x, v, c='r', label='v')
plt.legend()
plt.show()
```

```python
implicit_solution = np.zeros((x.size, 2))
implicit_solution[0] = np.array([u0, v0])
```

```python
for i in range(x.size - 1):
    A = np.array([[a, b], [c, d]])
    E = np.eye(2)
    mat = np.linalg.inv(2 * E / h - A)
    implicit_solution[i + 1] = mat.dot(f(*implicit_solution[i])) + implicit_solution[i]
impl_u = implicit_solution[:, 0]
impl_v = implicit_solution[:, 1]
plt.scatter(x, impl_u, c='b', label='u')
plt.scatter(x, impl_v, c='r', label='v')
plt.legend()
plt.show()
```
