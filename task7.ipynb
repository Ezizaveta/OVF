```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
initial_prey = 5
initial_predator = 5.1
```

```python
def RungeKutt_2(t,u,f, h, alpha=0.75):
    for i in range(len(t)-1):
        u[i+1] = u[i] + h*((1-alpha)*f(t[i], *u[i])+ alpha*f(t[i] + h/(2*alpha), *u[i] + h/(2*alpha)*f(t[i], *u[i])))
    return u
```

```python
N = 1000
a = 10
b = 2
c = 2
d = 10
h = 1 / 1000
T = 15
t = np.linspace(0, T, N)
u = np.empty((len(t), 2))
u[0] = np.array([initial_prey, initial_predator])
```

```python
f = lambda t, x, y: np.array([a*x - b*x*y, c*x*y - d*y])
u = RungeKutt_2(t, u, f, h)
x = u[:, 0]
y = u[:, 1]
plt.plot(x, y)
plt.show()
```
