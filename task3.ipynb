```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
def trapezoid_method(f, a, b, N):
    xi = a
    xj = a + (b - a) / N
    h = (b - a) / N
    I = 0
    for i in range(N):
        I += (f(xi) + f(xj)) * h / 2
        xi += h
        xj += h
    return I
```

```python
def simpsons_method(f, a, b, N):
    xi = a
    xj = a + (b - a) / N
    h = (b - a) / N
    I = 0
    for i in range(N):
        I += (f(xi) + 4 * f((xj + xi) / 2) + f(xj)) * h / 6
        xi += h
        xj += h
    return I
```

```python
f = lambda x: 1 / (1 + x ** 2)
```

```python
def integrate(f, a, b, N):    
    trapezoidal_solutions = np.zeros_like(N, dtype=np.float)
    simpsons_solutions = np.zeros_like(N, dtype=np.float)
    j = 0
    for i in N:
        trapezoidal_solutions[j] = trapezoid_method(f, -1, 1, i)
        simpsons_solutions[j] = simpsons_method(f, -1, 1, i)
        j += 1
        
    return trapezoidal_solutions, simpsons_solutions
    
```

```python
N = np.array([2 ** i for i in range(11)])
trapezoidal_solutions, simpsons_solutions = integrate(f, -1, 1, N)

plt.plot(N, np.abs(trapezoidal_solutions), label='trapezoid', marker='*')
plt.plot(N, np.abs(simpsons_solutions), label='simpsons', marker='*')
plt.grid()
plt.xticks(N)
plt.legend()
plt.xscale('log')
plt.yscale('log')
plt.show()
```
